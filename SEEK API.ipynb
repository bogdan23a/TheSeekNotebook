#                                                Using SEEK API for writing reports

```python
from SEEK import SEEK, auth
```

### Authentication

```python
auth = auth()
```

### Searching

```python
assay = SEEK(auth)
assay.request(type="assays", id="576")
```

```python
assay.print()
```

### Browsing

```python
search = SEEK(auth)
search.search()
```

```python
for item in search.find("liver"):
    if hasattr(item.data.attributes, 'description') and item.data.attributes.description != None:
        print(item.data.attributes.description + "\n")
```

### Flexibility in Browsing

#### Step by step guide to browsing
#### Step 1: Build a list to send to the search engine

You can do that either by using the API Search

```python
s = SEEK(auth)
s.demoSearch()
```

```python
s.createRequestList()
print(str(len(s.requestList)) + " requests found")
```

#### or ...

by even adding each item in the list. Each item has the format { 'id' : 223 , 'type' : 'assays' }.

```python
requestList = []
requestList.append({'id': '12', 'type':'assays'})
requestList.append({'id': '12', 'type':'data_file'})
requestList.append({'id': '12', 'type':'investigation'})
requestList.append({'id': '13', 'type':'assays'})
requestList.append({'id': '14', 'type':'assays'})
requestList.append({'id': '15', 'type':'assays'})
requestList.append({'id': '16', 'type':'assays'})
requestList.append({'id': '17', 'type':'assays'})
requestList.append({'id': '18', 'type':'assays'})
requestList.append({'id': '19', 'type':'assays'})
```

#### Step 2: Run the parallel search engine 

```python
ps = SEEK(auth)
ps.parallelRequest(requestList, requestPerThread=1)
```

#### Step 3: Build the relationship list

```python
totalNumberRelationships = ps.createRelationshipList()
print(str(totalNumberRelationships) + " relations found")
```

#### Step 4: Run the parallel search engine  

```python
PS = SEEK(auth)
PS.parallelRequest(ps.relationshipList, requestPerThread=10)
```

#### Step 5: Substitute the relationships in the search results

```python
ps.substituteRelationships(PS.requestList, totalNumberRelationships)
```

#### Step 6: Print the results

```python
for request in ps.requestList:

    request.print()
    print('\n________________________________________________________\n')
```
